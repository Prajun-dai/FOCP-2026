```python
def to_binary(number):
    return bin(number)[2:]


# Test program
n = int(input("Enter a positive integer: "))
print("Binary representation:", to_binary(n))
```

```python
def factors(number):
    result = []

    for i in range(1, number + 1):
        if number % i == 0:
            result.append(i)

    return result


# Test program
n = int(input("Enter an integer: "))
print("Factors:", factors(n))
```

```python
def is_prime(number):
    if number <= 1:
        return False

    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False

    return True


# Test program
n = int(input("Enter an integer: "))
print("Is prime:", is_prime(n))
```

```python
def simple_encrypt(message):
    no_spaces = message.replace(" ", "")
    return no_spaces[::-1]


# Test program
text = input("Enter a message: ")
print("Encrypted message:", simple_encrypt(text))
```

```python
import random
import string

def hide_message(message):
    interval = random.randint(2, 20)
    message = message.replace(" ", "")
    encrypted = ""

    for char in message:
        encrypted += char
        for _ in range(interval - 1):
            encrypted += random.choice(string.ascii_lowercase)

    return encrypted, interval


# Test program
msg = input("Enter a message: ")
encrypted_msg, spacing = hide_message(msg)
print("Encrypted message:", encrypted_msg)
print("Interval used:", spacing)
```

```python
def reveal_message(encrypted, interval):
    message = ""

    for i in range(0, len(encrypted), interval):
        message += encrypted[i]

    return message


# Test program
hidden, gap = hide_message("send cheese")
print("Hidden:", hidden)
print("Decrypted:", reveal_message(hidden, gap))
```
