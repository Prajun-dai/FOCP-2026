```python
name = input("Hello, what is your name? ")
print(f"Hello, {name}. Good to meet you!")
```

```python
celsius = float(input("Enter a temperature in Celsius: "))
fahrenheit = (celsius * 9 / 5) + 32

print(f"{celsius}C is equivalent to {fahrenheit}F.")
```

```python
students = int(input("How many students? "))
group_size = int(input("Required group size? "))

groups = students // group_size
leftover = students % group_size

if leftover == 1:
    print(f"There will be {groups} groups with {leftover} student left over.")
else:
    print(f"There will be {groups} groups with {leftover} students left over.")
```

```python
sweets = int(input("How many sweets are there? "))
pupils = int(input("How many pupils are present? "))

each = sweets // pupils
leftover = sweets % pupils

print(f"Each pupil gets {each} sweets.")
print(f"There will be {leftover} sweets left over.")
```
