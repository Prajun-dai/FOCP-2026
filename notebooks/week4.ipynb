```python
def is_valid_score(value):
    return 0 <= value <= 100


# Test program
num = int(input("Enter an integer: "))
print(is_valid_score(num))
```

```python
def count_case(text):
    upper = 0
    lower = 0

    for char in text:
        if char.isupper():
            upper += 1
        elif char.islower():
            lower += 1

    return upper, lower


# Test program
message = input("Enter a string: ")
u, l = count_case(message)
print("Uppercase letters:", u)
print("Lowercase letters:", l)
```

```python
def remove_last_character(text):
    if len(text) <= 1:
        return text
    return text[:-1]


# Test program
user_input = input("Enter a string: ")
print(remove_last_character(user_input))
```

```python
def c_to_f(celsius):
    return (celsius * 9 / 5) + 32


def f_to_c(fahrenheit):
    return (fahrenheit - 32) * 5 / 9


# Test program
c = float(input("Enter temperature in C: "))
print("In Fahrenheit:", c_to_f(c))

f = float(input("Enter temperature in F: "))
print("In Centigrade:", f_to_c(f))
```

```python
import statistics

def c_to_f(celsius):
    return (celsius * 9 / 5) + 32


temps = []

for i in range(6):
    temp = input("Enter temperature (e.g. 25C): ")
    value = float(temp[:-1])
    temps.append(c_to_f(value))

print("Maximum:", max(temps))
print("Minimum:", min(temps))
print("Mean:", statistics.mean(temps))
```

```python
import statistics

def c_to_f(celsius):
    return (celsius * 9 / 5) + 32


temps = []

while True:
    temp = input("Enter temperature (press Enter to finish): ")
    if temp == "":
        break
    value = float(temp[:-1])
    temps.append(c_to_f(value))

if temps:
    print("Maximum:", max(temps))
    print("Minimum:", min(temps))
    print("Mean:", statistics.mean(temps))
else:
    print("No temperatures entered.")
```
