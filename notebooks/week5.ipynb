```python
import sys
import urllib.request
import urllib.error

if len(sys.argv) < 2:
    print("Please provide a URL.")
    sys.exit()

url = sys.argv[1]

try:
    response = urllib.request.urlopen(url)
    print("Website is working. HTTP status code:", response.getcode())
except urllib.error.URLError:
    print("Website is not reachable.")
```

```python
import sys

# Get temperature values from command line if available
if len(sys.argv) > 1:
    temps_input = sys.argv[1:]
else:
    # Jupyter-safe: ask user interactively
    temps_input = input("Enter temperatures separated by spaces: ").split()

# Handle case where no values were provided
if not temps_input:
    print("No temperature values provided.")
else:
    try:
        # Convert all to floats
        temps = [float(t) for t in temps_input]

        maximum = max(temps)
        minimum = min(temps)
        mean = sum(temps) / len(temps)

        print(f"Max: {maximum}")
        print(f"Min: {minimum}")
        print(f"Mean: {mean}")

    except ValueError:
        print("Invalid input! Please enter numbers only.")
```

```python
import sys
import shutil
import os

# Get filename from command line or ask interactively
if len(sys.argv) > 1:
    filename = sys.argv[1]
else:
    filename = input("Enter the filename to backup: ")

# Check if file exists
if not os.path.isfile(filename):
    print(f"File '{filename}' does not exist.")
else:
    # Create backup name
    backup_name = filename + ".bak"

    try:
        shutil.copyfile(filename, backup_name)
        print(f"Backup created: {backup_name}")
    except Exception as e:
        print("Error creating backup:", e)
```
