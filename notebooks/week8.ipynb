```python
import sys
import os

# Get filenames from command line or input
if len(sys.argv) >= 3:
    file1 = sys.argv[1]
    file2 = sys.argv[2]
else:
    file1 = input("Enter first filename: ")
    file2 = input("Enter second filename: ")

# Check file existence
if not os.path.isfile(file1) or not os.path.isfile(file2):
    print("One or both files do not exist.")
else:
    with open(file1, "r") as f1, open(file2, "r") as f2:
        content1 = f1.read()
        content2 = f2.read()
    
    if content1 == content2:
        print("Files are the same.")
    else:
        print("Files are different.")
```

```python
import sys
import os

# Get search string and filename
if len(sys.argv) >= 3:
    pattern = sys.argv[1]
    filename = sys.argv[2]
else:
    pattern = input("Enter string to search for: ")
    filename = input("Enter filename: ")

# Check if file exists
if not os.path.isfile(filename):
    print(f"File '{filename}' does not exist.")
else:
    with open(filename, "r") as f:
        for line in f:
            if pattern in line:
                print(line.rstrip())
```

```python
import sys
import os

# Get filename from command line or input
if len(sys.argv) > 1:
    filename = sys.argv[1]
else:
    filename = input("Enter filename to count: ")

# Check if file exists
if not os.path.isfile(filename):
    print(f"File '{filename}' does not exist.")
else:
    with open(filename, "r") as f:
        text = f.read()

    # Count lines, words, and characters
    num_lines = text.count("\n") + 1 if text else 0
    num_words = len(text.split())
    num_chars = len(text)

    # Print results
    print(f"Lines: {num_lines}")
    print(f"Words: {num_words}")
    print(f"Characters: {num_chars}")
```

```python
import sys
import os
import string

# Load dictionary (simplified for cross-platform)
# You can download a word list from: https://github.com/dwyl/english-words
dict_file = "words.txt"  # Put your word list here

# Check if dictionary exists
if not os.path.isfile(dict_file):
    print(f"Dictionary file '{dict_file}' not found!")
    sys.exit(1)

# Load valid words
with open(dict_file, "r") as f:
    valid_words = set(word.strip().lower() for word in f)

# Get filename from command line or input
if len(sys.argv) > 1:
    filename = sys.argv[1]
else:
    filename = input("Enter filename to spell-check: ")

# Check if the file exists
if not os.path.isfile(filename):
    print(f"File '{filename}' does not exist.")
    sys.exit(1)

# Read file content
with open(filename, "r") as f:
    text = f.read()

# Remove punctuation
translator = str.maketrans('', '', string.punctuation)
clean_text = text.translate(translator)

# Find unknown words
words_in_file = set(word.lower() for word in clean_text.split())
unknown_words = words_in_file - valid_words

# Print unknown words
if unknown_words:
    print("Unknown words:")
    for w in sorted(unknown_words):
        print(w)
else:
    print("No unknown words found.")
```

```python
import sys
import urllib.request
import urllib.error

# Get URL safely (works in terminal and Jupyter)
if len(sys.argv) >= 2:
    url = sys.argv[1]
else:
    url = input("Enter a URL: ")

# Add scheme if missing
if not url.startswith(("http://", "https://")):
    url = "http://" + url

try:
    response = urllib.request.urlopen(url)
    print("Website is working.")
    print("HTTP status code:", response.getcode())

except urllib.error.HTTPError as e:
    print("Website returned an error.")
    print("HTTP status code:", e.code)

except urllib.error.URLError:
    print("Website is not reachable.")
```
